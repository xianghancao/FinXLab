# networkx的最短路径算法

```python
# version 1.0
from tqdm import tqdm
import networkx
import numpy as np

def networkx_dijkstra(m, source=None, target=None):
    G = networkx.Graph()
    for i in m.columns:
        G.add_node(i)
    for i in range(len(m.columns)):
        for j in range(i+1, len(m.columns)):
            if not np.isnan(m.iloc[i,j]):
                G.add_edge(m.columns[i], m.columns[j], weight=m.iloc[i,j])
    
    if source is not None and target is not None:
        return nx.dijkstra_path_length(G, source=source, target=target)
    
    p = m*np.nan
    for i in range(len(m.columns)):
        for j in range(i, len(m.columns)):
            try:
                p.iloc[i, j] = networkx.dijkstra_path_length(G, source=m.columns[i], target=m.columns[j])
            except networkx.exception.NetworkXNoPath:
                p.iloc[i, j] = np.nan
            p.iloc[j, i] = p.iloc[i, j]
    return p


def networkx_plot(G):
    import matplotlib.pyplot as plt
    plt.rcParams['font.sans-serif'] = 'SimHei'  # 设置字体为SimHei
    plt.rcParams['axes.unicode_minus'] = False  # 解决负号“-”显示异常
    pos = networkx.shell_layout(G)
    networkx.draw(G, pos, 
            with_labels=True,
            node_color='white',
            edge_color='red',
           node_size=400,
           alpha=0.5)
    pylab.title('network', fontsize=15)
    pylab.show()
```
