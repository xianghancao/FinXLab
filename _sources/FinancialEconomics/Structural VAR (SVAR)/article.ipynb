# **Structural VAR (SVAR): Overview**
 

First: 2024.12.27 Updated:2024.12.27 Version: 0.0.1


A Structural Vector Autoregression (SVAR) model is an extension of the traditional Vector Autoregression (VAR) model that incorporates **theoretical or structural restrictions** to identify specific, interpretable economic or financial shocks. Unlike the basic VAR, which provides reduced-form relationships, SVAR explicitly identifies and disentangles **causal mechanisms** in a system of interrelated variables.

---

### **Purpose of SVAR**

1. **Identifying Structural Shocks:**
   - Structural shocks are specific disturbances (e.g., monetary policy changes, geopolitical events) that are assumed to be orthogonal (uncorrelated) and interpretable.
   - SVAR helps isolate the impact of these shocks on the system.

2. **Relaxing Cholesky Limitations:**
   - The Cholesky decomposition used in basic VAR assumes a particular ordering of variables, which may not align with theory. SVAR overcomes this limitation by imposing meaningful restrictions.

3. **Economic Interpretability:**
   - SVAR ensures that impulse response functions (IRFs) and forecast error variance decompositions (FEVDs) align with economic theory or domain-specific knowledge.

---

### **Theoretical Framework**

#### **1. Reduced-Form VAR**
The reduced-form VAR is expressed as:
$$
Y_t = A_1 Y_{t-1} + \ldots + A_p Y_{t-p} + u_t
$$
Where:
- $ Y_t $: Vector of endogenous variables.
- $ A_1, \ldots, A_p $: Coefficient matrices.
- $ u_t $: Vector of reduced-form residuals (correlated shocks).

#### **2. Structural Form**
The structural form of VAR introduces contemporaneous relationships between variables:
$$
B Y_t = B_1 Y_{t-1} + \ldots + B_p Y_{t-p} + \epsilon_t
$$
Where:
- $ B $: Matrix of contemporaneous relationships.
- $ \epsilon_t $: Vector of structural shocks (assumed orthogonal).

The relationship between reduced-form residuals ($ u_t $) and structural shocks ($ \epsilon_t $) is:
$$
u_t = B^{-1} \epsilon_t
$$

#### **3. Identification Problem**
To identify $ \epsilon_t $, additional restrictions on $ B $ are required because the number of free parameters in $ B $ exceeds the number of equations.

---

### **Identification Methods**

1. **Zero Restrictions:**
   - Some variables do not affect others contemporaneously.
   - Example: Geopolitical risk may not be contemporaneously influenced by financial market returns.

2. **Sign Restrictions:**
   - Specify the expected direction of the impact of a shock.
   - Example: A geopolitical shock increases risk premia and decreases stock prices.

3. **Long-Run Restrictions:**
   - Assume that certain shocks have no long-term effect on some variables.
   - Example: A temporary monetary policy shock may not affect long-run GDP levels.

4. **Short-Run Restrictions:**
   - Constrain contemporaneous effects based on economic theory.

---

### **Application of SVAR**

#### **Use Case: Geopolitical Events and Financial Markets**
- Variables: Geopolitical Risk Index (GPR), stock returns, bond yields.
- Goal: Identify and measure the impact of geopolitical shocks on financial markets.

#### **Steps in Analysis**
1. **Fit a Reduced-Form VAR Model:**
   - Estimate the relationships among variables without imposing structural assumptions.

2. **Apply Structural Restrictions:**
   - Define the $ B $ matrix based on theory (e.g., GPR influences stock and bond markets contemporaneously, but not vice versa).

3. **Conduct Impulse Response Analysis (IRA):**
   - Measure the dynamic effects of a geopolitical shock on financial variables.

4. **Perform Variance Decomposition:**
   - Quantify the contribution of structural shocks to the variability of each variable.
