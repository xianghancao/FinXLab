```r
# 读取 CSV 文件，假设文件名为 "index_data.csv"
data <- read.csv("datasets/log_final_df.csv", header = TRUE)
```

```r
# 假设第一列为日期列，将其转换为日期格式
data[, 1] <- as.Date(data[, 1], format = "%Y-%m-%d")  # 根据文件中的日期格式修改格式字符串
# 查看数据的前几行
head(data)
```

```r
library(tvReg)
VAR.fit <- vars::VAR(usmacro, p = 1, type = "const")
```

```r
coefficients(VAR.fit)
```
